# Cleaning DF

```python
def clean_braces(brace_str):
    if isinstance(brace_str, str):
        brace_str = brace_str.replace('[', '')
        brace_str = brace_str.replace(')', '')
    return brace_str
```

```python
df['age'] = df['age'].apply(clean_braces)
df['age'].unique()
```

```python
df['weight'] = df['weight'].apply(clean_braces)
df['weight'].unique()
```

```python

counts = df.apply(lambda col: col.astype(str).str.count(r'\?').sum())


print(counts)
```

# Ans Questions 1

**Show relation between people's age, race, gender with diabet med and non diabate med**

```python
df_ques1=df[['patient_nbr','race','gender','age','diabetesMed']]
df_ques1
```

```python
df_ques1['race'].unique()
```

```python

counts = df_ques1.apply(lambda col: col.astype(str).str.count(r'\?').sum())


print(counts)
```

```python

df_ques1 = df_ques1.replace('?', np.nan)
df_ques1
```

```python

counts = df_ques1.apply(lambda col: col.astype(str).str.count(r'\?').sum())

print(counts)
```

```python
df_ques1.info()
```

```python

df_ques1 = df_ques1.dropna()
df_ques1
```

```python
age_diabetesMed_counts = df_ques1.groupby(['age', 'diabetesMed']).size().unstack(fill_value=0)

age_diabetesMed_counts.plot(kind='bar')
plt.show()
```

So age 70-80 group has most encounters for diabatese problem

```python
gender_diabetesMed_counts = df_ques1.groupby(['gender', 'diabetesMed']).size().unstack(fill_value=0)

gender_diabetesMed_counts.plot(kind='bar')
plt.show()
```

Female has most tendency to get dibetese
